```
# Coletando os dados da internet e exibindo apenas as 5 primeiras linhas
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
dados = pd.read_csv(uri)
dados.head()
```

```
# Renomear o nome das tabelas

mapa = {
    "home" : "principal",
    "how_it_works" : "como_funciona",
    "contact" : "contato",
    "bought" : "comprou"
}
dados = dados.rename(columns= mapa)
```

```
# Separando os dados de entrada do resultado
x = dados[["principal","como_funciona","contato"]]
y = dados["comprou"]
dados.head()
```

```
# Verificando o formato dos dados (99 linhas e 4 colunas)
dados.shape
```

```
# Separando os dados que serão utilizados como teste(24) e treino(75)

treino_x = x[:75]
treino_y = y[:75]
teste_x = x[75:]
teste_y = y[75:]

print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))
```

```
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acuracia foi de %.2f%%" % acuracia)
```
